```python
df = pd.read_csv("weather_data.csv", parse_dates=["day"]) #string to date type Timestamp
type(df.day[0]) 
```

```python
#df.fillna(0, inplace = True) #null replace with 0
df
```

```python
df.fillna(0, inplace = False)
df
```

```python
df.temperature.median()
```

```python
df.windspeed.median()
```

```python
df.fillna({
    'temperature' : df.temperature.median(),
    'windspeed' : df.windspeed.median(),
    'event': 'No Event'
})
```

```python
df
```

```python
df.fillna(method='ffill') #sets previous columns value to next one hence day was set to index to track forward fill
```

```python
df.fillna(method='bfill') 
```
